```python
def boxHistPlot(housing_data_1, figsize=(16, 5)):
    plt.figure(figsize=figsize)
    plt.subplot(121)
    sns.boxplot(housing_data_1)
    plt.subplot(122)
    sns.distplot(housing_data_1)
```
